# 7.5 2 SUBPLOTS

benimEksenler dizi oldu ve dizilerde öyle metod yok 

o yüzden diziyi loop da değerlendireceğiz

```python
(benimFigure,benimEKsenler)=plt.subplots(nrows=1,ncols=2)

for eksen in benimEKsenler:
    eksen.plot(yas_list_np,yas_list_np,"b--")
    eksen.set_xlabel("x ekseni ")

plt.tight_layout() # İÇ İÇE GİRMELERİ ENGELLEME
# benimFigure # ÇİZMEZSE DİYE
```

```python
numpyDizisi1=np.arange(0,10)
numpyDizisi2=numpyDizisi1**2
```

legend() birden fazla grafiği renklendirir ve adlabndırır. 

loc=1,2,... yazarak yerini değiştirebiliriz.

dpi=100 grafiği büyütmeye/küçültmeye yarar dpi(dots per inch)

```python
yeniFigur=plt.figure(dpi=100)

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numpDizisi **2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpDizisi **3")

yeniEksen.legend(loc=1)
```

savefig(dosya_Adi,vesaireler) Dosyayı kaydeder (bu dosyanın bulunduğu yere kaydeder)

```python
yeniFigur.savefig("yeniFigur.png",dpi=200)
```
