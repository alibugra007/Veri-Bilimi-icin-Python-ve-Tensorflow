```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
npDizi=np.linspace(0,10,20)
npDizi_2=npDizi**2
```

HTML deki renk kodlarını burada da kullanabiliriz.

#060505   formatında

alpha=0.4  <1 olur. 1 normal halidir. Transparan (şeffaf/mat) görünüm için.

```python
(benimFigur,benimEksen)=plt.subplots()
benimEksen.plot(npDizi,npDizi_2,"#060505", alpha=0.5)
benimEksen.plot(npDizi_2,npDizi,"b")
```

linewidth=5  Çizgi kalınlığı ayarlama

linestyle="--" hat stili ayarlama  

marker="o" işaretleyici x lerin olduğu yere grafikte işaret koyuyor

markerfacecolor="red" o işaretlerin rengini değiştirir

```python
(benimFigur_2,benimEksen_2)=plt.subplots()
benimEksen_2.plot(npDizi,npDizi,color="red",linewidth=5)
benimEksen_2.plot(npDizi,npDizi+2,color="blue",linewidth=1)
benimEksen_2.plot(npDizi,npDizi+4,color="black",linewidth=1, linestyle="--")
benimEksen_2.plot(npDizi,npDizi-2,color="yellow",linewidth=5, linestyle="-", marker="o",markerfacecolor="blue")
```
